```python
import requests
from bs4 import BeautifulSoup

url = "https://www.mbishop.com.tw/Article/Detail/79781"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
script = soup.find_all('script')
# script[-3]
```

```python
import re
import html
import re

js_data = script[-3].string

# 解碼 HTML 實體
decoded_html = html.unescape(js_data)

# 移除 HTML 標籤
clean_text = re.sub(r'<[^>]*>', '', decoded_html)

# print(clean_text)

introduction_match = re.search(r'Introduction:"(.*?)",\s*MainPicURL', clean_text, re.S)
# introduction_match = re.sub(r'img { max-width:100%; height:auto; } ', ' ', introduction_match)
subtitle_match = re.search(r'SubTitle:\s*`([^`]+)`', clean_text)
clean_text = re.sub(r'SubTitle', ' ', clean_text)
title_match = re.search(r' Title:\s*`([^`]+)`', clean_text)

if introduction_match and subtitle_match and title_match:
    introduction_text = introduction_match.group(1)

    subtitle_text = subtitle_match.group(1)
    title_text = title_match.group(1)
    
    # 去除引号内的 HTML 实体和多余空白
    introduction_text = introduction_text.strip()
    introduction_text = re.sub(r'\\n', '\n', introduction_text)
    introduction_text = re.sub(r'\s+', ' ', introduction_text)
    
    print("Title:")
    print(title_text)
    print("\nSubtitle:")
    print(subtitle_text)
    print("\nIntroduction:")
    print(introduction_text)
```

```python
import pandas as pd
import gspread
from google.oauth2.service_account import Credentials
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
```

```python
combined_text = f"{title_text}\n\n{subtitle_text}\n\n{introduction_text}"

def chunk_text(text, chunk_size=500):
    words = text.split()
    for i in range(0, len(words), chunk_size):
        yield ' '.join(words[i:i + chunk_size])

# 加载预训练的BERT模型和tokenizer
tokenizer = AutoTokenizer.from_pretrained("hfl/chinese-bert-wwm-ext")
model = AutoModelForTokenClassification.from_pretrained("hfl/chinese-bert-wwm-ext")

# 使用pipeline进行NER
nlp = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")

# Function to process text in chunks
def process_ner_in_chunks(text, chunk_size=512):
    words = text.split()
    entities = []
    for i in range(0, len(words), chunk_size):
        chunk = ' '.join(words[i:i + chunk_size])
        ner_results = nlp(chunk)
        for entity in ner_results:
            if 'entity_group' in entity and 'word' in entity:
                entities.append((entity['word'], entity['entity_group'], entity['score']))
    return entities

# # Process the combined text with the NER model
entities = process_ner_in_chunks(combined_text)



# ner_results = []
# for chunk in chunk_text(combined_text):
#     ner_results.extend(nlp(chunk))

# # ner_results = nlp(combined_text)

# # 解析NER结果
# entities = []
# for entity in ner_results:
#     if 'entity_group' in entity and 'word' in entity:
#         entities.append((entity['word'], entity['entity_group'], entity['score']))

# 创建DataFrame
df = pd.DataFrame(entities, columns=["Word", "Entity", "Score"])
df
```
